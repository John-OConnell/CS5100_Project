```python
data = data[["Date", "Close"]]
print(data.head())
```

```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
plt.figure(figsize=(15, 10))
plt.plot(data["Date"], data["Close"])
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(data["Close"], 
                            model='multiplicative', period = 30)
fig = plt.figure()  
fig = result.plot()  
fig.set_size_inches(15, 10)
```

```python
pd.plotting.autocorrelation_plot(data["Close"])
```

```python
from statsmodels.graphics.tsaplots import plot_pacf
plot_pacf(data["Close"], lags = 100)
```

```python
train_data = data['Close'][:int(len(data)*0.95)]
test_data = data['Close'][int(len(data)*0.95):]
p, d, q = 5, 1, 2
# from statsmodels.tsa.arima_model import ARIMA
import statsmodels.api as sm
model = sm.tsa.ARIMA(data["Close"], order=(p,d,q))
# model = ARIMA(data.Close, order=(1,0,0))
# model = ARIMA(data["Close"], order=(p,d,q))  
fitted = model.fit()  
print(fitted.summary())
```

```python
predictions = fitted.predict()
print(predictions)
```

```python
predictions = fitted.predict(len(data), len(data)+25)
print(predictions)
data["Close"].plot(legend=True, label="Training Data", figsize=(15, 10))
predictions.plot(legend=True, label="Predictions")
```

```python
import statsmodels.api as sm
import warnings
model=sm.tsa.statespace.SARIMAX(train_data,
                                order=(p, d, q),
                                seasonal_order=(p, d, q, 12))
model=model.fit()
print(model.summary())
```

```python
predictions = model.predict(len(train_data), len(train_data)+len(test_data))
print(predictions)
```

```python
data["Close"].plot(legend=True, label="Training Data", figsize=(15, 10))
predictions.plot(legend=True, label="Predictions")
```
